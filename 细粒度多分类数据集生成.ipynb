```python
import os
import shutil


def get_file_list(folder: str):
    """
    获取文件夹下的全部文件名称
    :param folder:
    :return:
    """
    result = []
    if os.path.isfile(folder):
        return result
    for item in os.listdir(folder):
        if os.path.isdir(item):
            continue
        result.append(item)
    return result


def get_folder_list(folder: str):
    """
    获取文件夹下的全部文件夹名称
    :param folder:
    :return:
    """
    result = []
    for item in os.listdir(folder):
        if os.path.isfile(item):
            print("{} is file!".format(item))
            continue
        result.append(item)
    return result


def copy_file(source, target):
    if not os.path.exists(os.path.dirname(target)):
        os.makedirs(os.path.dirname(target))
    shutil.copy(source, target)


def train_val_pred_split(files, val_size, pred_size):
    """
    将传入的文件列表分为三分，训练，验证，测试
    :param files:
    :param val_size:
    :param pred_size:
    :return: val,pred,train
    """
    total = len(files)
    if val_size < 1 or pred_size < 1:
        val_num = int(total * val_size)
        pred_num = int(total * pred_size)
        return files[:val_num], files[val_num:val_num + pred_num], files[val_num + pred_num:]
    else:
        return files[:val_size], files[val_size:val_size + pred_size], files[val_size + pred_size:]
```

```python
val_num = 10
pred_num = 10
source_root = r"I:\wei\dataset\VisualPhish\trusted_list"
target_root = r"E:\dataset"
target_train = r"I:\wei\NTS-Net\dataset\train"
target_val = r"I:\wei\NTS-Net\dataset\val"
target_pred = r"I:\wei\NTS-Net\dataset\pred"
```

```python
# 用于生成数据集
for i, folder in enumerate(get_folder_list(source_root)):
    files = get_file_list(os.path.join(source_root, folder))
    # print("{} has picture {}".format(folder,len(files)))
    if(len(files)<val_num*2):
        continue
    val, pred,train = train_val_pred_split(files, val_num, pred_num)
    print("{} start!".format(folder))
    for item in train:
        copy_file(os.path.join(source_root, folder, item),
                  os.path.join(target_train, folder, item))
    for item in val:
        copy_file(os.path.join(source_root, folder, item),
                  os.path.join(target_val, folder, item))

    for item in pred:
        copy_file(os.path.join(source_root, folder, item),
                  os.path.join(target_pred, folder, item))
```
